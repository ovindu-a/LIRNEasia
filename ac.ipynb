```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
ac_roster = pd.read_csv('w1_ac_roster.csv')
consumption_data = pd.read_csv('non_smart_meter_data.csv')

# Merge the datasets on household_ID
merged_data = pd.merge(ac_roster, consumption_data, on='household_ID')

# Define continuous and categorical variables
continuous_vars = [
    'wattage_of_the_ac', 'btu_of_the_ac', 
    'no_of_hours_ac_was_on_during_daytime_last_week', 
    'no_of_hours_ac_was_on_during_night_last_week'
]

categorical_vars = [
    'room_ID', 'ac_ID', 'type_of_the_ac', 
    'is_the_ac_inverter_or_not', 'is_room_fully_sealed'
]

# Create scatter plots for continuous variables
for var in continuous_vars:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=var, y='consumption', data=merged_data)
    plt.title(f'Consumption vs {var}')
    plt.xlabel(var)
    plt.ylabel('Consumption')
    plt.yscale('linear')
    plt.tight_layout()
    plt.show()

# Create box plots for categorical variables
for var in categorical_vars:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=var, y='consumption', data=merged_data)
    plt.title(f'Consumption vs {var}')
    plt.xticks(rotation=45)
    plt.yscale('linear')
    plt.tight_layout()
    plt.show()
```
